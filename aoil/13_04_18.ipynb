# Convenient regrouping
Given $z_1, \dots , z_n \in \mathbb{Z}$, $|z_i| \leq W$.
What is the largest and smallest value of an expression of the form
$$
    p_{A,B} = \left(\sum_{i \in A} z_i\right)\left(\sum_{i \in B} z_i\right)
$$
where $A \cap B = \emptyset$ and $A \cup B = \{1, \dots , n\}$? 

Devise a polynomial procedure (in n) to compute them.

```julia
v = [5, 4, 4, 3, 2, 2, 2]
```

```julia
# Brute force solution generated by ChatGPT
function brute_force_min_max(z::Vector{Int})
    n = length(z)
    S = sum(z)
    # Initialize minimum and maximum values
    min_val = typemax(Int)
    max_val = typemin(Int)
    
    # There are 2^n ways to partition the set. Each partition corresponds
    # to a subset A (and B = complement of A).
    for mask in 0:(2^n - 1)
        sA = 0
        # Construct the subset A corresponding to the bitmask
        for i in 1:n
            if mask & (1 << (i-1)) != 0
                sA += z[i]
            end
        end
        # Compute the product p_{A,B}
        p = sA * (S - sA)
        min_val = min(min_val, p)
        max_val = max(max_val, p)
    end
    return min_val, max_val
end

# Example usage:
min_p, max_p = brute_force_min_max(v)
println("Minimum value: ", min_p)
println("Maximum value: ", max_p)
```
