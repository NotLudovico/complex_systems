```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 10_000  # Dimension of the matrix
var = 1  # Variance for the normal distribution

# Generate a random symmetric matrix
X = np.random.normal(0, var, size=(N, N))
M = (X + X.T) / np.sqrt(2 * N)

X_2 = np.random.normal(0, var, size=(N, N))
M_2 = (X_2 + X_2.T) / np.sqrt(2 * N)

# Calculate eigenvalues
e1 = np.linalg.eigvals(M)
e2 = np.linalg.eigvals(M_2)
e_sum = np.linalg.eigvals(M+M_2)

```

```python
plt.figure(figsize=(12, 8))

plt.hist(e1, bins=50, alpha=0.6, label="Eigenvalues of M1", density=True)
plt.hist(e2, bins=50, alpha=0.6, label="Eigenvalues of M2", density=True)
plt.hist(
    e_sum, bins=50, alpha=0.6, label="Eigenvalues of M1 + M2", density=True
)

plt.title("Eigenvalue Distributions")
plt.xlabel("Eigenvalue")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
