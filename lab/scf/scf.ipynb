# Ecut Convergence

```python
ecut_list = [10,15,20,25,30,33,35]
energies = [
    -6.1699135548E+01,
    -6.2054101486E+01,
    -6.2209492301E+01,
    -6.2265600019E+01,
    -6.2281613669E+01,
    -6.2284338044E+01,
    -6.2285061430E+01
]

from abipy.tools.plotting import ConvergenceAnalyzer

ca = ConvergenceAnalyzer.from_xy_label_vals(
    "ecut (Ha)", ecut_list, "E/natom (eV)", energies, tols=1e-3
)

print(ca)
ca.plot()
```
